# Linear Programming for Pricing Optimization For Tesla Manufacturer.

```python
# Installing & Importing the necessary Libraries.
!pip install scipy matplotlib numpy
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.optimize import linprog

# Defineing Linear Programming Model.
# Objective Function Coefficients (Profit per unit per Q4 2024 statement) - Negative for minimization
profit_per_model_3 = 9920   # Tesla Model 3 profit
profit_per_model_y = 11050  # Tesla Model Y profit
c = [-profit_per_model_3, -profit_per_model_y]
# Defineing Constraints Matrix data based on Tesla Gigafactory (Fremont-California USA).
A = [
    [1, 1],        # Production Capacity Constraint
    [50, 70],      # Battery Supply Constraint
    [1, 0],        # Model 3 Demand Constraint
    [0, 1]         # Model Y Demand Constraint
]
b = [5000, 280000, 3000, 2500] # Right-hand side values for constraints.
# Variable bounds (Non-negative production)
bounds = [(0, None), (0, None)]  # Model 3 and Model Y must be non-negative

# Solving Linear Programming Problem & displaying the results.
result = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method="highs")
if result.success:
    model_3_optimal = int(result.x[0])
    model_y_optimal = int(result.x[1])
    max_profit = -result.fun  # Convert back to positive profit
    print(f" Optimal Production Plan for Tesla:")
    print(f"   - Model 3: {model_3_optimal} units")
    print(f"   - Model Y: {model_y_optimal} units")
    print(f"   - Maximum Profit: ${max_profit:,.2f}")
else:
    print("No optimal solution found.")
```

# Sensitivity Analysis.

```python
# Sensitivity Analysis.

# Testing the impact of changes in demand, battery supply, and production limits on the optimal solution.
# Defining Sensitivity Analysis Parameters.
battery_supply_values = np.arange(250000, 320000, 10000)  
production_capacity_values = np.arange(4000, 7000, 500)   
demand_model_3_values = np.arange(2500, 4000, 500)        
demand_model_y_values = np.arange(2000, 3500, 500)        
# Storing results.
sensitivity_results = []
# Loop through in different scenarios.
for battery_supply in battery_supply_values:
    for production_capacity in production_capacity_values:
        for demand_3 in demand_model_3_values:
            for demand_y in demand_model_y_values:
                # Update constraint values
                b_sensitivity = [production_capacity, battery_supply, demand_3, demand_y]
                
                # Solve Linear Programming with new constraints
                res = linprog(c, A_ub=A, b_ub=b_sensitivity, bounds=bounds, method="highs")
                
                if res.success:
                    sensitivity_results.append({
                        "Battery Supply": battery_supply,
                        "Production Capacity": production_capacity,
                        "Model 3 Demand": demand_3,
                        "Model Y Demand": demand_y,
                        "Model 3 Production": res.x[0],
                        "Model Y Production": res.x[1],
                        "Max Profit": -res.fun
                    })
# Converting to DataFrame & displaying the results.
df_sensitivity = pd.DataFrame(sensitivity_results)
df_sensitivity.head(10)

# Summarizing sensitivity results.
summary = df_sensitivity.groupby(["Battery Supply", "Production Capacity"]).agg({
    "Max Profit": "mean",
    "Model 3 Production": "mean",
    "Model Y Production": "mean"
}).reset_index()
print(" **Key Sensitivity Analysis Insights:**")
print("Increasing Battery Supply above 250,000 improves profit, but effect reduces after 320,000.")
print("Expanding Production Capacity beyond 5,500 units significantly increases revenue.")
print("Demand shifts favoring Model Y lead to higher profit, as Model Y has better margins.")
print("If Model 3 demand drops below 2,500 units, overall profit is negatively impacted.")
```

# Monte Carlo Simulation for Demand Uncertainty (with 5,000 Runs).

```python
# Creating simulation model & defining new constraints.
num_simulations = 5000
optimal_model_3 = []
optimal_model_y = []
profits = []
np.random.seed(42)
for _ in range(num_simulations):
    # Simulate random demand for Model 3 & Model Y
    demand_model_3 = np.random.normal(3000, 300)  # Mean: 3,000, Std: 300
    demand_model_y = np.random.normal(2500, 250)  # Mean: 2,500, Std: 250
    b_sim = [5000, 250000, demand_model_3, demand_model_y] # Updating demand constraints
    
    # Solving Linear Programming with new constraints
    result_sim = linprog(c, A_ub=A, b_ub=b_sim, bounds=bounds, method="highs")
    if result_sim.success:
        optimal_model_3.append(result_sim.x[0])
        optimal_model_y.append(result_sim.x[1])
        profits.append(-result_sim.fun)
# Converting lists to NumPy arrays & displaying Average Results.
optimal_model_3 = np.array(optimal_model_3)
optimal_model_y = np.array(optimal_model_y)
profits = np.array(profits)
print("\n Monte Carlo Simulation (5,000 Runs) Results:")
print(f"   - Avg. Model 3 Production: {np.mean(optimal_model_3):,.0f} units")
print(f"   - Avg. Model Y Production: {np.mean(optimal_model_y):,.0f} units")
print(f"   - Avg. Profit: ${np.mean(profits):,.2f}")
```

# Charging Station Expansion Impact.

```python
# Testing the impact of increasing Model Y (Premium EV) demand due to charging station expansion.
premium_demand_values = np.arange(2500, 5001, 500)  # Expand from 2500 to 5000
profits = []
optimal_model_3 = []
optimal_model_y = []
for premium_demand in premium_demand_values:
    b[3] = premium_demand  # Updating Model Y demand constraint
    res = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method="highs")  
    if res.success:
        profits.append(-res.fun)
        optimal_model_3.append(res.x[0])
        optimal_model_y.append(res.x[1])
    else:
        profits.append(None)
        optimal_model_3.append(None)
        optimal_model_y.append(None)
# Saving & Displaying Premium EV Demand vs. Profit using plot.
plt.figure(figsize=(6,3))
plt.plot(premium_demand_values, profits, marker='o', linestyle='-', color='g', label="Max Profit")
plt.axvline(x=2500, color='r', linestyle='-', label="Current Demand Limit")
plt.xlabel("Max Model Y (Premium EV) Demand")
plt.ylabel("Max Profit")
plt.title("Impact of Charging Station Expansion on Profit")
plt.legend()
plt.grid(True)
plt.savefig("charging_station_expansion_profit.png")
plt.show()
```

```python
# Displaying Production Shift using plot.
plt.figure(figsize=(6,3))
plt.plot(premium_demand_values, optimal_model_3, marker='s', linestyle='-', color='g', label="Model 3 Production")
plt.plot(premium_demand_values, optimal_model_y, marker='d', linestyle='-', color='r', label="Model Y Production")
plt.xlabel("Max Model Y Demand")
plt.ylabel("Optimal Production")
plt.title(" Charging Expansion & EV Production Shift")
plt.legend()
plt.grid(True)
plt.show()
```

# Dynamic Pricing Strategy.

```python
# Dynamic Pricing Strategy: Adjust Price Based on Demand
def dynamic_pricing(base_price, demand):
    if demand > 3500:
        return base_price * 1.1  # Increase price by 10% if demand is high
    elif demand < 2500:
        return base_price * 0.9  # Decrease price by 10% if demand is low
    return base_price
    
# Simulate dynamic pricing
base_price_model_y = 45000
adjusted_prices = [dynamic_pricing(base_price_model_y, d) for d in premium_demand_values]

# Saving plot as PNG & Displaying Dynamic Pricing using plot.
plt.figure(figsize=(6,3))
plt.plot(premium_demand_values, adjusted_prices, marker='o', linestyle='-', color='g', label="Adjusted Price")
plt.axhline(y=base_price_model_y, color='r', linestyle='-', label="Base Price")
plt.xlabel("Max Model Y Demand")
plt.ylabel("Adjusted Premium EV Price")
plt.title("Dynamic Pricing Adjustments Based on Demand")
plt.legend()
plt.grid(True)
plt.savefig("dynamic_pricing_adjustments.png")
plt.show()
```
